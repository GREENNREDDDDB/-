**【过拟合与欠拟合】**
**【欠拟合】**：欠拟合（under-fitting）也称为欠学习，它的直观表现是算法训练得到的模型在训练集上表现差，没有学到数据的规律。
**【过拟合】**：过拟合（over-fitting）也称为过学习，它的直观表现是算法在训练集上表现好，但在测试集上表现不好，泛化性能差（拟合了训练样本集中的噪声）。
