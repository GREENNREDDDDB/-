>12. **touch 文档名 ：创建文档**
## 3、重定向 Redirection 
#### 一般unix命令都从terminal输入再从terminal输出，【例如在terminal输入cat命令，后面你向terminal输入什么，terminal就向你输出什么直到你按下Ctrl+d停止】，而重定向则是：
>1. **用“命令 > 目标文档” ：将命令的输出给到文档里【cat > list1 ,之后向terminal输入什么，输出都将记录到list1里】**
>2. **用“命令 > 目标文档” ：将命令的输出都追加到文档里（上一个会把文档清空）**
>3. **用“cat file1 file2 > file3”:将file1和file2的内容连接起来给到一个新文档file3内（文档会被自动创建）**
>4. **用“命令 < 目标文档” ：将目标文档的内容输入到命令里**
>5. **用“sort < file1 > file2 : 将file1的内容给到sort排序，后输出到file2里（sort ：对输入进行数字或字母排序）**
>6. **用管道pipe”命令1 | 命令2“ ：将命令1的结果送到命令2处理【who | sort 将用户排序】**

>4. **用 ”apropos 对命令的形容“  来获得记不太清楚的操作的命令名**
